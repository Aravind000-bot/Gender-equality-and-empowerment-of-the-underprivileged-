````python
import streamlit as st

# Function for codebot responses
def codebot_response(user_input):
    user_input = user_input.lower()

    # Gender equality and empowerment resources
    if "gender equality" in user_input:
        return ("Gender equality means equal rights and opportunities for all genders.\n"
                "You can read more on UN Women: https://www.unwomen.org/")

    elif "empowerment" in user_input:
        return ("Empowerment is about increasing the strength and confidence of individuals, "
                "particularly marginalized groups. It can include gaining access to resources like education, "
                "economic independence, and legal rights.")

    elif "resources" in user_input:
        return ("Available resources include:\n"
                "1. Local women’s shelters\n"
                "2. Education programs\n"
                "3. Healthcare services\n"
                "4. Legal aid services")

    # Basic coding help
    elif "python" in user_input:
        return ("Python is a versatile programming language. Here’s an example of a Python function:\n"
                "```python\n"
                "def greet(name):\n"
                "    return f'Hello, {name}!'\n"
                "print(greet('Alice'))\n"
                "```\n"
                "This function prints a greeting message.")

    elif "javascript" in user_input:
        return ("JavaScript is essential for web development. Here's an example of a JavaScript function:\n"
                "```javascript\n"
                "function greet(name) {\n"
                "    return 'Hello, ' + name + '!';\n"
                "}\n"
                "console.log(greet('Alice'));\n"
                "```\n"
                "This function prints a greeting message in the console.")

    elif "html" in user_input:
        return ("HTML is the foundation of web pages. Here's a simple HTML structure:\n"
                "```html\n"
                "<!DOCTYPE html>\n"
                "<html>\n"
                "<head>\n"
                "    <title>Simple Page</title>\n"
                "</head>\n"
                "<body>\n"
                "    <h1>Welcome to My Page</h1>\n"
                "</body>\n"
                "</html>\n"
                "```\n"
                "This code creates a basic HTML page with a title and a heading.")

    else:
        return "I can help with gender equality resources or programming questions. Ask me about coding in Python, JavaScript, or HTML, or about gender equality and empowerment."

# Streamlit UI
st.title("Codebot for Gender Equality and Empowerment")
st.write("This chatbot provides coding advice as well as resources on gender equality and empowerment. Ask a question!")

# Input from user
user_input = st.text_input("You: ", "Type your message here...")

# Display the response
if user_input:
    response = codebot_response(user_input)
    st.write(f"Codebot: {response}")
````
